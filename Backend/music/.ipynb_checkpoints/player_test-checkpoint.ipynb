```python
import vlc
import pafy
import time
# url of the video
url = "https://www.youtube.com/watch?v=38-ncylS_KM"
video = pafy.new(url)  #設定URL
best = video.getbest()  #選擇最高畫質播放
# print(best.resolution, best.extension)  #可以看選擇到的畫質為多少
media = vlc.MediaPlayer(best.url)  #創建播放器
media.audio_set_delay(2000)
# 開始串流播放
media.play()
time.sleep(5)
```
